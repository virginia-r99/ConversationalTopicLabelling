```python
# Import libraries
from transformers import pipeline, Conversation
import pandas as pd
import re
```

```python
# Define conversational models names
conv_models = ["facebook/blenderbot-400M-distill",\
    "microsoft/DialoGPT-medium","microsoft/GODEL-v1_1-base-seq2seq",\
    "microsoft/DialoGPT-small","facebook/blenderbot_small-90M",\
    "facebook/blenderbot-3B","microsoft/DialoGPT-large",\
    "microsoft/GODEL-v1_1-large-seq2seq"]

conv_names = ["blenderbot-400M-distill",\
    "DialoGPT-medium","GODEL-v1_1-base-seq2seq",\
    "DialoGPT-small","blenderbot_small-90M",\
    "blenderbot-3B","DialoGPT-large",\
    "GODEL-v1_1-large-seq2seq"]
```

```python
# Define QA models names
qa_models = ["distilbert-base-cased-distilled-squad","bert-large-uncased-whole-word-masking-finetuned-squad",\
    "deepset/roberta-base-squad2","deepset/minilm-uncased-squad2",\
    "deepset/bert-large-uncased-whole-word-masking-squad2","deepset/roberta-base-squad2-covid",\
    "deepset/bert-base-cased-squad2","monologg/koelectra-small-v2-distilled-korquad-384",\
    "sultan/BioM-ELECTRA-Large-SQuAD2","deepset/roberta-large-squad2"]

qa_names = ["distilbert-base-cased-distilled-squad","bert-large-uncased-whole-word-masking-finetuned-squad",\
    "roberta-base-squad2","minilm-uncased-squad2",\
    "bert-large-uncased-whole-word-masking-squad2","roberta-base-squad2-covid",\
    "bert-base-cased-squad2","koelectra-small-v2-distilled-korquad-384",\
    "BioM-ELECTRA-Large-SQuAD2","roberta-large-squad2"]
```

```python
# Define the numbers of words to compute
nval = [1,2,3,4,5,7,9]
```

```python
# For each QA model (by index)
for qa in range(0,len(qa_models)):    
        # Generate QA model
        qamodel = pipeline("question-answering",model=qa_models[qa])
        # For each number of words
        for n in nval:
            # For each conversational model (by index)
            for c in range(0,len(conv_models)):
                # Define dataframe to save the QA results
                topics_results = pd.DataFrame(columns=['Real topic','Conv answer','Prediction'])
                # Read topics dataset
                topics = pd.read_json("Results/q1/topic/"+str(n)+"/topics_conv_results_"+str(n)+"_"+conv_names[c]+".json")
                topics = topics.T
                # For each topic (by index)
                for i in range(0,len(topics.index)): 
                    # Extract the conversational answer
                    response = topics['Conv answer'].values[i]
                    # Generate the QA answer given the conv. answer
                    answ = qamodel(question="What is the topic that contains the words?", context=response)
                    
                    # Show the QA answer
                    print("The topic is "+answ['answer'])
                    
                    # Save topic results to topics_results
                    topics_results.loc[len(topics_results.index)] = [topics['Real topic'].values[i],response,answ['answer']]
                    
                # Save topics_results to json file
                topics_results.to_json("Results/q1/"+str(n)+"/topics_results_"+str(n)+"_"+conv_names[c]+"_"+qa_names[qa]+".json",orient="index")
                print("Results saved "+str(n)+": "+conv_names[c]+" "+qa_names[qa])
```
