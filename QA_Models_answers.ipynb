```python
# Import libraries
from transformers import pipeline, Conversation
import pandas as pd
import re
```

1. Define the conversational models, QA models, and number of topic words and saving path that will be used along the QA answers' generation.

```python
# Define conversational models names
conv_models = ["facebook/blenderbot-400M-distill", "microsoft/DialoGPT-medium"]

conv_names = ["blenderbot-400M-distill", "DialoGPT-medium"]
```

```python
# Define QA models names
qa_models = ["deepset/deberta-v3-large-squad2", "deepset/deberta-v3-base-squad2",\
    "deepset/xlm-roberta-large-squad2", "deepset/bert-large-uncased-whole-word-masking-squad2",\
     "deepset/roberta-base-squad2-distilled"]

qa_names = ["deberta-v3-large-squad2", "deberta-v3-base-squad2",\
    "xlm-roberta-large-squad2", "bert-large-uncased-whole-word-masking-squad2",\
     "roberta-base-squad2-distilled"]
```

```python
# Define the number of words of the description to ask in array format where each element is a number of words to use
n = [1,2,3,4,5,7,9]
```

2. For each QA model, each number of topic words and each conversational model, generate the QA answer from the Conversational model where the topic label is contained and save the results in a JSON file named "topics_results_n_conv_qa.json" where _n_ is the number of words, _conv_ is the conversational model's name and _qa_ is the QA model's name in each case. The JSON file consists on a list of the topic ground truth labels and the QA answer recieved in order.

```python
# For each QA model (by index)
for qa in range(0,len(qa_models)):    
        # Inicialize QA model
        qamodel = pipeline("question-answering",model=qa_models[qa])
        # For each number of words
        for n_i in n:
            # Define path to save QA answers at
            save_path = "Results/q1/"+str(n_i)+"/
            # For each conversational model (by index)
            for c in range(0,len(conv_models)):
                # Define dataframe to save the QA results
                topics_results = pd.DataFrame(columns=['Real topic','Conv answer','Prediction'])
                # Read topics dataset
                topics = pd.read_json("Results/q1/topic/"+str(n_i)+"/topics_conv_results_"+str(n_i)+"_"+conv_names[c]+".json")
                topics = topics.T
                # For each topic (by index)
                for i in range(0,len(topics.index)): 
                    # Extract the conversational answer
                    response = topics['Conv answer'].values[i]
                    # Generate the QA answer given the conv. answer
                    answ = qamodel(question="What is the topic that contains the words?", context=response)
                    
                    # Show the QA answer
                    print("The topic is "+answ['answer'])
                    
                    # Save topic results to topics_results
                    topics_results.loc[len(topics_results.index)] = [topics['Real topic'].values[i],response,answ['answer']]
                    
                # Save topics_results to json file
                topics_results.to_json(save_path+"topics_results_"+str(n_i)+"_"+conv_names[c]+"_"+qa_names[qa]+".json",orient="index")
                print("Results saved "+str(n_i)+": "+conv_names[c]+" "+qa_names[qa])
```
